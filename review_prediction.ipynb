```python
from tkinter import *
import tkinter as tk
from tkinter import messagebox
from tkinter import ttk
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
#from sklearn.naive_bayes import BernoulliNB
#from sklearn.naive_bayes import GaussianNB
import string
from tkinter import filedialog


words=stopwords.words("english")
words.remove('not')
words.remove("don't")
words.remove("didn't")
words.remove("wasn't")
words.remove("haven't")
words.remove("hasn't")
words.remove("hadn't")
words.remove("weren't")
words.remove("wouldn't")
words.remove("couldn't")
words.remove("shouldn't")
ps=PorterStemmer()
def clean_text(msg):
    msg=re.sub(f"[{string.punctuation}]","",msg)
    msg=msg.lower()
    list_wrd=word_tokenize(msg)
    temp_list=[]
    for w in list_wrd:
        if(w not in words):
            temp_list.append(ps.stem(w))
    return " ".join(temp_list)

#please change the storage path for your training and testing data according to your system.
df=pd.read_csv("Restaurant_Reviews.txt",delimiter='\t')
df.Review=df.Review.map(clean_text)
cv=CountVectorizer(binary=False,ngram_range=(1,2))
clf=MultinomialNB()
X=cv.fit_transform(df.Review).toarray()
y=df.Liked
clf.fit(X,y)

window=Tk()
window.geometry("850x470")
window.configure(bg='black')
window.title('Review prediction system')
window.resizable(height=False,width=False)

label_main=Label(window,text='Review Prediction System',font=('serif',40,'bold italic'),fg='white',bg='black')
label_main.pack()

def mainscreen():
    frame=Frame(window,bg='black')
    frame.place(x=0,y=100,relwidth=1,relheight=1)
    label_name=Label(frame,text='by Aayush Mishra',font=('serif',20,'bold'),bg='black',fg='firebrick2')
    label_name.place(x=580,y=320)
    label_username=Label(frame,text='Username : ',font=('serif',20,'bold'),bg='black',fg='white')
    label_username.place(x=200,y=80)
    entry_username=Entry(frame,font=('',15,'bold'))
    entry_username.place(x=420,y=80)
    entry_username.focus()
    label_password=Label(frame,text='Password : ',font=('serif',20,'bold'),bg='black',fg='white')
    label_password.place(x=200,y=150)
    entry_password=Entry(frame,font=('',15,'bold'),show='*')
    entry_password.place(x=420,y=150)
    login_button=Button(frame,command=lambda:login(frame,entry_username,entry_password),text='Login',font=('serif',15,'bold'),bd=3,width=9)
    login_button.place(x=350,y=270)
def login(frame,i1,i2):
    user=i1.get()
    pas=i2.get()
    if(len(user)==0 or len(pas)==0):
        messagebox.showwarning('Warning','Username or Password can not be empty, please enter again. ')
    else:
        if(user=='admin' and pas=='admin'):
            messagebox.showinfo('Success','Welcome to review prediction.')
            frame.destroy()
            selection_screen()
        else:
            messagebox.showerror('Error','username or Password')
def selection_screen():
    frame=Frame(window,bg='black')
    frame.place(x=0,y=100,relwidth=1,relheight=1)
    btn_single_feedback=Button(frame,command=lambda:single_feedback(frame),text='Single Feedback',font=('serif',15,'bold'),bd=5,width=18)
    btn_single_feedback.place(x=310,y=100)
    btn_bulk_feedback=Button(frame,command=lambda:bulk_feedback(frame),text='Bulk Feedback',font=('serif',15,'bold'),bd=5,width=18)
    btn_bulk_feedback.place(x=310,y=200)
def single_feedback(frame):
    frame.destroy()
    frame=Frame(window,bg='black')
    frame.place(x=0,y=100,relwidth=1,relheight=1)
    label_main_2=Label(frame,text='Single Review Prediction',font=('times',25,'bold italic'),bg='black',fg='white')
    label_main_2.place(x=270,y=0)
    label_feedback=Label(frame,text='Review : ',font=('serif',20,'bold'),bg='black',fg='white')
    label_feedback.place(x=200,y=80)
    label_feedback=Entry(frame,font=('',15,'bold'))
    label_feedback.place(x=420,y=80)
    predict_button=Button(frame,command=lambda:single_prediction_main(frame,label_feedback,label_predicted),text='Predict',font=('serif',15,'bold'),bd=3,width=9)
    predict_button.place(x=350,y=170)
    back_button=Button(frame,command=lambda:back_btn(frame),text='back',font=('serif',15,'bold'),bd=5,width=7)
    back_button.place(x=50,y=300)
    label_predicted=Label(frame,text='prediction Result',font=('times',20,'bold'),bg='black',fg='white')
    label_predicted.place(x=600,y=300)
def single_prediction_main(frame,e1,e2):
    test=clean_text(e1.get())
    test_X=cv.transform([test])
    pred=clf.predict(test_X.toarray())
    if(pred[0]==0):
        e2.configure(text="negeative feedback",fg='red')
    if(pred[0]==1):
        e2.configure(text="positive feedback",fg='blue')
def back_btn(frame):
    frame.destroy()
    selection_screen()
def back_btn_bulk(frame):
    frame.destroy()
    selection_screen()
def bulk_feedback(frame):
    frame.destroy()
    frame=Frame(window,bg='black')
    frame.place(x=0,y=100,relwidth=1,relheight=1)
    label_main_2=Label(frame,text='Bulk Review Prediction',font=('times',25,'bold italic'),bg='black',fg='white')
    label_main_2.place(x=270,y=0)
    back_button=Button(frame,command=lambda:back_btn_bulk(frame),text='back',font=('serif',15,'bold'),bd=5,width=7)
    back_button.place(x=20,y=300)
    lbl_word_1=Label(frame,text='Select File : ',font=('',20,'bold'),bg='black',fg='white')
    lbl_word_1.place(x=160,y=60)
    entry_file_name=Entry(frame,font=('',15,'bold'),bd=5)
    entry_file_name.place(x=320,y=60)
    btn_browse_file=Button(frame,command=lambda:browse_files(frame,entry_file_name),text='Browse',font=('',15,'bold'),height=1)
    btn_browse_file.place(x=500,y=100)
    label_save=Label(frame,text='Predict File And Save As',font=('',20,'bold'),bg='black',fg='white')
    label_save.place(x=270,y=200)
    entry_save_file_name=Entry(frame,font=('',15,'bold'),bd=5)
    entry_save_file_name.place(x=200,y=250)
    btn_save_file=Button(frame,command=lambda:predict_and_save_files(frame,entry_file_name,entry_save_file_name,label_complete),text='predict and save file',font=('',15,'bold'),bd=5)
    btn_save_file.place(x=540,y=250)
    label_complete=Label(frame,text='***',font=('',20,'bold'),bg='black',fg='black')
    label_complete.place(x=270,y=320)
def browse_files(frame,e1):
    dict_path=filedialog.askopenfilename(initialdir="/",title="Open File",filetypes=(("Text Files", "*.txt"), ("All Files", "*.*")))
    e1.insert(0,dict_path)
    e1.config(state='readonly')
def predict_and_save_files(frame,e1,e2,e3):
    path=e1.get()
    save_as=e2.get()
    #the test document must have a data in the given format otherwise prediction may give error.
    df=pd.read_csv(path,delimiter='\t')
    list_b=[]
    for i in df.Review:
        test=clean_text(i)
        test_X=cv.transform([test])
        pred=clf.predict(test_X.toarray())
        if(pred[0]==0):
            list_b.append((i,'negative feedback'))
        if(pred[0]==1):
            list_b.append((i,'positive feedback'))
        df_rand=pd.DataFrame(list_b,columns=['Review','Feedback'] )
        df_rand.to_csv(f'{save_as}.txt',index=False,index_label=False,sep='\t')
        e3.configure(text='file saved successfully',fg='blue')

    
        
mainscreen()
window.mainloop()

```
